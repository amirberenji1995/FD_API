```python
import face_recognition
from PIL import Image
import cv2
import matplotlib.pyplot as plt
```

```python
def face_detector(add):
    image = face_recognition.load_image_file(add)
    face_locations = face_recognition.face_locations(image)
    face_rois = []
    for face_location in face_locations:
        top, right, bottom, left = face_location
        face_image = image[top:bottom, left:right]
        face_rois.append(Image.fromarray(face_image).convert('LA').resize((48,48)))
        image = cv2.rectangle(image, (left, top), (right, bottom), (255,0 , 0), 25)
    return image, face_rois
```
